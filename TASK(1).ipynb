```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
file_path = "API_SP.POP.TOTL_DS2_en_csv_v2_38144.csv"
df = pd.read_csv(file_path, skiprows=4)
```

```python
year = "2020"
```

```python
countries_df = df[~df["Country Name"].str.contains("World|income|IDA|IBRD|OECD|Euro|Africa|Asia|Europe|America|Caribbean|Arab", case=False, na=False)]
```

```python
pop_data = countries_df[["Country Name", year]].dropna()
```

```python
top10 = pop_data.sort_values(by=year, ascending=False).head(10)
```

```python
plt.figure(figsize=(10,6))
plt.bar(top10["Country Name"], top10[year] / 1e9)  # convert to billions
plt.title(f"Top 10 Most Populated Countries in {year}")
plt.ylabel("Population (Billions)")
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.hist(pop_data[year] / 1e6, bins=30, edgecolor="black")  # population in millions
plt.title(f"Distribution of Country Populations in {year}")
plt.xlabel("Population (Millions)")
plt.ylabel("Number of Countries")
plt.show()
```
